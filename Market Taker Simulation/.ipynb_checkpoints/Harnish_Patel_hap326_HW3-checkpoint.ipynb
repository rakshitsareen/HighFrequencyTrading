```python
executedpricelist = []
executedprice = 0
returns = []
ret = 0
TTElist = []
tte = 0
SLlist = []
sl = 'No'
TTEtriggreslist = []
ttetriggred = 'No'
exewithinTTElist = []
exeintte = 'Yes'
for i in range(df3.shape[0]):
    executedpricelist.append(executedprice)
    returns.append(ret)
    TTElist.append(tte)
    SLlist.append(sl)
    TTEtriggreslist.append(ttetriggred)
    exewithinTTElist.append(exeintte)
    if df3.iloc[i]['signal'] == '1':
        #print('buy')
        t1 = df3.iloc[i]['order datetime']
        orderedbid = df3['Bid'][i]
        ask_array = df2.loc[df2.index>=t1]['Ask'].values
        time_array = df2.loc[df2.index>=t1].index
        executedprice = 0
        ret = 0
        tte = 0
        sl = 'No'
        ttetriggred = 'No'
        exeintte = 'Yes'
        stoploss = 0.5*(df3['Ask'][i] - df3['Bid'][i]) #ask_array.max() - ask_array[0]
        maxTTE = np.timedelta64(10, 's') #np.timedelta64(df2.index[-1] - t1)
        for j in range(ask_array.size):
            if ask_array[j]<=orderedbid:
                executedprice = ask_array[j]
                ret = (df3['Mid'][i] - executedprice)/df3['Mid'][i]
                tte = time_array[j] - t1
                break
            if (time_array[j] - t1)>= maxTTE:
                #print('TTE occured', i)
                executedprice = ask_array[j]
                ret = (df3['Mid'][i] - executedprice)/df3['Mid'][i]
                tte = time_array[j] - t1
                ttetriggred = 'Yes'
                exeintte = 'No'
                #print('TTE lenght buy', time_array[j] - t1)
                break
            if (ask_array[j] - df3['Ask'][i])>=stoploss:
                #print('stop loss occured', i)
                #print('buy time delta', time_array[j] - t1)
                executedprice = ask_array[j]
                ret = (df3['Mid'][i] - executedprice)/df3['Mid'][i]
                tte = time_array[j] - t1
                sl = 'Yes'
                break

    else:
        #print('sell')
        t1 = df3.iloc[i]['order datetime']
        orderedask = df3['Ask'][i]
        bid_array = df2.loc[df2.index>=t1]['Bid'].values
        time_array = df2.loc[df2.index>=t1].index
        executedprice = 0
        ret = 0
        tte = 0
        sl = 'No'
        ttetriggred = 'No'
        exeintte = 'Yes'
        stoploss = 0.5*(df3['Ask'][i] - df3['Bid'][i]) #ask_array.max() - ask_array[0]
        maxTTE = np.timedelta64(10, 's') #np.timedelta64(df2.index[-1] - t1)
        for j in range(bid_array.size):
            if bid_array[j]>=orderedask:
                executedprice = bid_array[j]
                ret = (executedprice - df3['Mid'][i])/df3['Mid'][i]
                tte = time_array[j] - t1
                break
            if (time_array[j] - t1)>= maxTTE:
                #print('TTE occured', i)
                executedprice = bid_array[j]
                ret = (executedprice - df3['Mid'][i])/df3['Mid'][i]
                tte = time_array[j] - t1
                ttetriggred = 'Yes'
                exeintte = 'No'
                #print('TTE lenght sell', time_array[j] - t1)
                break
            if (df3['Bid'][i] - bid_array[j])>=stoploss:
                #print('stop loss occured', i)
                #print('sell time delta', time_array[j] - t1)
                executedprice = bid_array[j]
                ret = (executedprice - df3['Mid'][i])/df3['Mid'][i]
                tte = time_array[j] - t1
                sl = 'Yes'
                break
                
executedpricelist.append(executedprice)
returns.append(ret)
TTElist.append(tte)
SLlist.append(sl)
TTEtriggreslist.append(ttetriggred)
exewithinTTElist.append(exeintte)
```

#### 1)	For each of your 3 approaches (MT, OMMSide and OMMMid), provide a recapitulating table which will show at the minimum the average and median execution PnL for each order (taken from Mid Market), average and median time to execution), the number of times when your Stop Loss was triggered, the number of times when your order was executed on Time Limit.

```python
recap_table = pd.DataFrame(columns = ['Approach', 'Average pnl', 'median pnl', 'minimum pnl', 'maximum pnl', 'average TTE(seconds)', 'median TTE(seconds)', 'Total numbers of SL triggered', 'Total numbers of times order executed on max TTE'])
```

```python
recap_table['Approach'] = np.array(['MT', 'OMMside', 'OMMmid'])

avgpnlT = []
medianpnlT = []
minpnlT = []
maxpnlT = []
avgtteT = []
mediantteT = []
tsltT = []
teontteT = []

avgpnlT.append(MT['returns'].mean())
medianpnlT.append(MT['returns'].median())
minpnlT.append(MT['returns'].min())
maxpnlT.append(MT['returns'].max())
avgtteT.append('NA')
mediantteT.append('NA')
tsltT.append('NA')
teontteT.append('NA')

avgpnlT.append(OMMside['returns'].mean())
medianpnlT.append(OMMside['returns'].median())
minpnlT.append(OMMside['returns'].min())
maxpnlT.append(OMMside['returns'].max())
avgtteT.append(OMMside['time to execution (seconds)'].mean()) #.total_seconds())
mediantteT.append(OMMside['time to execution (seconds)'].median()) #.total_seconds())
tsltT.append(OMMside['SL triggered'][OMMmid['SL triggered']=='Yes'].count())
teontteT.append(OMMside['SL triggered'][OMMmid['TTE triggered']=='Yes'].count())

avgpnlT.append(OMMmid['returns'].mean())
medianpnlT.append(OMMmid['returns'].median())
minpnlT.append(OMMmid['returns'].min())
maxpnlT.append(OMMmid['returns'].max())
avgtteT.append(OMMmid['time to execution (seconds)'].mean()) #.total_seconds())
mediantteT.append(OMMmid['time to execution (seconds)'].median()) #.total_seconds())
tsltT.append(OMMmid['SL triggered'][OMMmid['SL triggered']=='Yes'].count())
teontteT.append(OMMmid['SL triggered'][OMMmid['TTE triggered']=='Yes'].count())  

recap_table['Average pnl'] = np.array(avgpnlT)
recap_table['median pnl'] = np.array(medianpnlT)
recap_table['minimum pnl'] = np.array(minpnlT)
recap_table['maximum pnl'] = np.array(maxpnlT)
recap_table['average TTE(seconds)'] = np.array(avgtteT)
recap_table['median TTE(seconds)'] = np.array(mediantteT)
recap_table['Total numbers of SL triggered'] = np.array(tsltT)
recap_table['Total numbers of times order executed on max TTE'] = np.array(teontteT)
```

mean PnL for DIS alpha generators for all approach are in order OMMside>OMMmid>MT of from highest to lowest.
